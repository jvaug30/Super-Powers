```python
powers['Powers_LS'] = powers['Powers'].str.split(',')
```

```python
powers.head()
```

```python
powers_exploded = powers.explode('Powers_LS')[['hero_names', 'Powers_LS']]
powers_exploded
```

```python
cols_to_make = powers_exploded['Powers_LS'].dropna().unique()
cols_to_make
```

```python
for col in cols_to_make:
    powers[col] = powers['Powers'].str.contains(col)
powers.head()
```

```python
powers = powers.drop(columns=['Powers', 'Powers_LS'])
powers.head()
```

```python
hero_powers = pd.merge(hero, powers, on='hero_names')
```

```python
hero_powers.head()
```

```python
hero_powers['height_centimeters'] = hero_powers['height_centimeters'].astype(float)
```

```python
hero_powers['weight_kilograms'] = hero_powers['weight_kilograms'].astype(float)
```

```python
# Compare the average weight of super heros who have Super Speed to those who do not
hero_powers.groupby('Super Speed')['weight_kilograms'].mean()
```

```python
# What is the average height of heroes for each publisher
hero_powers.groupby('publisher')['height_centimeters'].mean()
```
